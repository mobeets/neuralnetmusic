```python
import os.path
import glob
import theano
```

```python
mididir = os.path.join(os.path.abspath(os.path.curdir), 'midifiles', 'train')
files = glob.glob(os.path.join(mididir, '*.mid'))
print files[0]
```

```python
'''r : (integer, integer) tuple
    Specifies the pitch range of the piano-roll in MIDI note numbers,
    including r[0] but not r[1], such that r[1]-r[0] is the number of
    visible units of the RBM at a given time step. The default (21,
    109) corresponds to the full range of piano (88 notes).
dt : float
    Sampling period when converting the MIDI files into piano-rolls, or
    equivalently the time difference between consecutive time steps.
    '''
r = (21, 109)
dt = 0.248958333333
dataset = [midiread(f, r, dt).piano_roll.astype(theano.config.floatX) for f in files[:10]]
dataset[0].T.shape
```

```python
ind = 3
fnm = files[ind]
snippet = dataset[ind]
midiwrite(fnm.replace('.mid', '_2.mid'), snippet, r=(12, 109), dt=128)
```
